# Divide a big CSV file into 2 or more files.

```python
import pandas as pd
```

## Define variables.

```python
csv_dir = ".\\Data\\"
csv_name = "Sample.csv"
csv_path = csv_dir + csv_name
csv_encoding = 'Shift_JISx0213'

# A number of files the CSV file will be divided into.
divCount = 3
```

## Read a big csv file as a DataFrame.
CSVファイルをDataFrameに読み込む。

  - The Sample.csv file contains data of 800,000 rows by 100 columns and one header row on the top. / Sample.csvには80万行ｘ100列のデータが含まれており、ヘッダが付いている。 

```python
df = pd.read_csv(csv_path, encoding=csv_encoding, header=0)
display(df.head(5))
display(df.tail(5))
print("The shape of the DataFrame --->", df.shape)
```

## Calculate the max rows in each CSV file.
分割後のCSVファイルに含まれる最大行数を計算する。

```python
maxRow = round(len(df) / divCount)
print("The number of max row is --->", maxRow)
```

## Divide the DataFrame by defined number then save as a new CSV file.
DataFrameを分割し、CSVファイルを生成。

```python
for i in range(divCount):
    start = i * maxRow
    end = (i + 1) * maxRow
    
    dfDiv = df[start : end]
    
    if i == divCount -1:
        end = len(df)
        dfDiv =  df[start :]

    print(  "-----  " + str(start+1) + '-' + str(end) + "  -----")
    print("The shape of the DataFrame --->", dfDiv.shape)
    
    divName = csv_path.replace('.csv', '-div' + str(i+1) + '_(' + str(start+1) + "-" + str(end) +').csv')
    print("The name of created CSV file --->", divName)
    
    dfDiv.to_csv(divName, encoding=csv_encoding, header=True, index= False)
    
print("---------------------------------------")
print("complete.")
```
